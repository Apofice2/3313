```python
#creation dataset
x , y = make_regression(n_samples= 1000, n_features= 2, noise= 10)
#dimensionnement de y
y = y.reshape(y.shape[0], 1)

plt.scatter(x[: ,0], y)
plt.scatter(x[: ,1], y ,c = 'r')
# Afficher le graphique
plt.show()

X = np.hstack((x , np.ones((x.shape[0], 1))))
print(y.shape, X.shape)
```

```python
theta = np.random.randn(3, 1)
theta.shape
theta
```

MODEL

```python
def model(x, theta):
  return x.dot(theta)

```

```python
def cost_fonction(X, y, theta):
  m = len(y)
  return (1/(2*m))*np.sum((model(X, theta) - y)**2)

cost_fonction(X, y, theta)
```

```python
def grad(x, y, theta):
  m = len(y)
  return (1/m) * x.T.dot(model(x, theta) - y)


grad(X, y, theta)
```

```python
def descente_grad(x, y, theta, learning_rate, nb_iteration):
  cost = np.zeros(nb_iteration)
  for i in range(0, nb_iteration):
    theta = theta - learning_rate * grad(x, y, theta)
    cost[i] = cost_fonction(x, y, theta)
  return theta, cost

```

```python
theta_final , cost = descente_grad(X, y, theta ,learning_rate = 0.1 ,nb_iteration= 1000)
theta_final
```

```python
prediction = model(X ,theta_final)
```

```python
plt.scatter(x[: ,0] ,y)
#plt.scatter(x[: ,1] ,y)
plt.plot(x , prediction ,c='r')
```
